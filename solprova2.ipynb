**Questão 1:**

Por definição $a^{n} = a\times a^{n-1}$ e $a^0=1$ (base da recursão).
Assim, uma função recursiva para o cálculo da poência de um número é dada por:
```
potencia(a, n){
    se n == 0 então
       Devolva 1
    caso contrário
       Devolva a*potencia(a, n-1)
}
```
Simulação: considere $a = 3$ e $n = 2$. A função deve devolver $3^2 = 9$.

Chamada da função com parâmetros: $a = 3$ e $n=2$:
```
potencia(3, 2)
n == 0 ? 2 = 0 ? Falso
Devolva 3 * potencia(3, 1)
            1 == 0 ? Falso
            Devolva 3 * potencia(3, 0)
                        0 == 0 ? Verdadeiro
                        Devolva 1
```
Assim, a função devolve $3\times 3\times 1 = 9$ com três chamadas da função recursiva potência.

Queremos determinar o campeão, ou seja, o número de inscrição do competidor que fez o menor tempo. Para isso precisamos percorrer as n linhas da matriz para determinar o menor tempo e armazenar o número de isncrição. Vamos considerar $M$ uma matriz $n\times 2$ que contém estas informações.

```
menor_tempo = M[0][1]
inscricao = M[0][0]
Para i = 1 até n, faça
   se menor_tempo > M[i][1] então
      menor_tempo = M[i][1]
      inscricao = i
Mostre "O campeao eh o competidor" inscricao
```

Simulação: Considere a matriz de resultados,
$$M = \left[\begin{array}{cc}
132 & 17.8\\
111 & 9.8 \\
125 & 10.9
\end{array}\right]$$

Número de linhas 3: $n=2$.
```
menor_tempo = 17.8
inscricao = 132
Para i = 1: 17.8 > 9.8 ? Verdadeiro, então
            menor_tempo = 9.8
            inscricao = 111
Para i = 2: 9.8 > 10.9 ? Falso

```
Assim, o algoritmo devolve: O campeão eh competidor 111.

**Questão 5:**

Esta função recebe uma matriz quadrada (tamanho $n\times n$) e devolve o melemento máximo em valor absoluto.

$$A=\left[\begin{array}{cc}
1 & -10 \\
9 & 2
\end{array}\right]$$

Simulação:

```
n = 2
m = 0
Para i = 0: (até 1)
   Para j = 0: (até 1)
       |A[0][0]| > m ? 1 > 0, Sim
       m = 1
   Para j = 1: (até 1)
       |A[0][1]| > m ? 10 > 1, Sim
       m = 10
Para i = 1: (até 1)
   Para j = 0: (até 1)
      |A[1][0]| > m ? 9 > 10, Não
   Para j = 1: (até 1)
      |A[1][1]| > m ? 2 > 10, Não
```
Assim a função devolve o valor $m=10$.
