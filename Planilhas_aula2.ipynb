````python
def media(P1, P2, P3, AP):
    return 0.8*(P1 + P2 + P3)/3 + 0.2*AP

df['media'] = media(df.P1, df.P2, df.P3, df.AP)
m = df['media'].sort_values(ascending=False)
print(m)

print(df)
````

**Visualização**

Exemplo de Gráfico da situação:

````python
def contabiliza_situacao(media):
    n = len(media)
    nr = 0
    nrec = 0
    na = 0
    for i in range(n):
        if(media[i] >= 6):
            na = na + 1
        elif(media[i] >= 3):
            nrec += 1
        else:
            nr += 1
    return [na, nrec, nr]

qtdade = contabiliza_situacao(df['media'])
situacao = ["Aprovados", "REC", "Reprovados"]
plt.pie(qtdade, labels = situacao, autopct = '%1.1f%%')
plt.title('Situação Estudantes')
plt.show()
````
