# Fundamentos

Nesta aula vamos estudar alguns conceitos fundamentais com o objetivo de escrever um primeiro programa na linguagem  *Python*.

**Tipos de dados, expressões e variáveis**

> Um computador manipula informação presente na memória. Estas informações se dividem em:



*   **Instruções**: comandos que determinam o funcionamento do computador e como os dados devem ser processados.
*   **Dados**: informações que devem ser manipuladas (processadas) pelo computador.

Os dados podem ser classificados em diferentes tipos.



*   **Numérico**: números inteiros (*int*), números reais (*float*), números complexos (*complex*).
*   **Literal**: Caracteres (*str - string*).
* **Lógico**: Verdadeiro ou Falso (*bool*).

Dados de tipos diferentes são armazenados de forma diferente. É importante definir o tipo de dado para definir a quantidade de memória utilizada para armazená-lo.

Experimente os comandos abaixo no *Python shell*:



```
type(1)
type(2.1)
type(False)
type(1 + 2j)
type('palavra')
```

Lembre-se que o *Python shell* é um ambiente interativo que espera um ENTER para executar o que foi digitado na linha de comando e imprime o resultado.  







**Variáveis:**

O computador armazena os dados em variáveis para poder manipular estes dados. Uma variável contém quatro informações:



*   **Nome**: usado durante o algoritmo quando se faz referência a vairável.
*   **Tipo**: o tipo de dado que a variável armazena.
*   **Conteúdo**: Valor armazenado na variável, compatível com o seu tipo.
*   **Endereço**: posição da memória em que a variável é armazenada.

Em algumas linguagens de programação, como C ou Fortran, é necessário declarar as variáveis, ou seja, definir o nome e o tipo de cada variável antes de usá-la. Exemplo de declaração em C:



```
int indice;
float nota;
char letra;
```
Depois de declarar as variáveis, ela não possuem nenhum conteúdo, apenas nome, tipo e endereço. Para que a variavel tenha um valor é necessário inicializar a variável, atribuindo-lhe um valor. Exemplo de inicialização na linguagem C:

```
indice = 1;
nota = 9.5;
letra = A;
```
Em *Python*, uma linguagem de programação fracamente tipada, ao atribuir um valor para a variável, a mesma recebe o conteúdo e o tipo.

O **operador** = é o operador de atribuição. O **Operador de atribuição** atribui valor a uma variável.

Experimente os comandos abaixo no Python Shell:

```
i = 1
type(1)
nota = 9.0
type(nota)
letra = "A"
type(letra)
```







**Expressões:**

Expressões são uma combinação d valores (dados, operandos) e operadores.



> **Expressão Aritmética**:


As expressões aritméticas são expressões matemáticas que envolvem números (inteiros, reais ou complexos) e os operadores de soma, divisão, subtração, multiplicação exponenciação, etc...

**Regras de precedência e associatividade**

$$\begin{array}{|c|c|c|}
\hline
\mbox{Operador} & \mbox{Descrição} & \mbox{Associatividade} \\
\hline
( ) & \mbox{Parênteses}  &  \mbox{da esquerda para a direita}\\
\hline
** & \mbox{Potência} & \mbox{da direita para a esquerda}\\
\hline
+, - & \mbox{positivo e negativo (unário)} & \mbox{da direita para a esquerda} \\
\hline
*, /, //, \% & \mbox{multiplicação, divisão, divisão inteira, resto} & \mbox{da esquerda para a direita} \\
\hline
+,- & \mbox{soma e subtração} & \mbox{da esquerda para a direita}\\
\hline
\end{array}$$

Experimente os comandos abaixo no Python Shell:

```
x = 3
y = 5
z = 3*x+y/2
type(z)
```
Qual é o resultado de $2 + 3*4$ e $(2 + 3)*4$?

O resultado da expressão $3/4*100$ é 75, pois como os operadores de divisão e multiplicação possuem o mesmo grau de precedência as operações são feitas da esquerda para a direita. Qual é o resultado de $3//4*100$?

Uma expressão aritmética válida (mesmo que longa) é “reduzida” a um valor.





> **Expressão Lógica**:

As expressões lógicas tem como resultado um valor lógico, **verdadeiro** ou **falso**.

Os **operadores relacionais** são usados para comparar valores numéricos e resultam em um valor lógico.

$$\begin{array}{|c|c|c|}
\hline
\mbox{Operador} & \mbox{Descrição} & \mbox{Exemplo} & \mbox{Resultado} \\
\hline
== & \mbox{Igual}  &  1==2 & \mbox{False}\\
\hline
!= & \mbox{Diferente} & 1!=2 & \mbox{True}\\
\hline
> & \mbox{Maior} & 2>2 & \mbox{False} \\
\hline
>= & \mbox{Maior ou igual} & 2>=2 & \mbox{True} \\
\hline
< & \mbox{Menor} & 2 < 3 & \mbox{True}\\
\hline
<= & \mbox{Menor ou igual} & 3 <=2 & \mbox{False} \\
\hline
\end{array}$$

Experimente os comandos abaixo no Python Shell:

```
a = 1
b = 2
a == 1
a != b
b < a
```

Uma comparação simples pode ser feita entre pares de expressões aritméticas, da forma:

< expressão aritmética 1> < operador relacional > < expressão aritmétixa 2 >

Por exemplo,

```
3**2 + 4*2 < 5*15//4
```



As **expressões lógicas** são compostas por valores lógicos e operadores lógicos. A tabela abaixo apresenta a ordem de precedência dos operadores lógicos usados em *Python*:

$$\begin{array}{|c|c|c|c|}
\hline
\mbox{Operador} & \mbox{Descrição} & \mbox{Exemplo} & \mbox{Resultado} \\
\hline \hline
\mbox{not} & \mbox{Negação}  &  \mbox{not  False} & \mbox{True}\\
\hline
\mbox{and} & \mbox{E} & \mbox{True and False} & \mbox{False}\\
\hline
\mbox{or} & \mbox{Ou} & \mbox{True or False} & \mbox{True} \\
\hline
\end{array}$$

A tabela-verdade é uma tabela que associa todos os possíveis valores lógicos dos operandos em expressões lógicas.

**Operador not**: troca o valor lógico da expressão.

$$\begin{array}{|c|c|}
\hline
\mbox{p} & \mbox{not p} \\
\hline
\mbox{True} & \mbox{False} \\
\hline
\mbox{False} & \mbox{True} \\
\hline
\end{array}$$

**Operador and**: resulta no valor True apenas quando os seus dois operandos possuem o valor True.

$$\begin{array}{|c|c|c|}
\hline
\mbox{p} & \mbox{q} & \mbox{p and q} \\
\hline
\mbox{True} & \mbox{True} &  \mbox{True}\\
\hline
\mbox{False} & \mbox{True} & \mbox{False} \\
\hline
\mbox{True} & \mbox{False} & \mbox{False}\\
\hline
\mbox{False} & \mbox{False} & \mbox{False} \\
\hline
\end{array}$$

**Operador or**: resulta no valor False apenas quando os seus dois operandos possuem o valor False.

$$\begin{array}{|c|c|c|}
\hline
\mbox{p} & \mbox{q} & \mbox{p or q} \\
\hline
\mbox{True} & \mbox{True} &  \mbox{True}\\
\hline
\mbox{False} & \mbox{True} & \mbox{True} \\
\hline
\mbox{True} & \mbox{False} & \mbox{True}\\
\hline
\mbox{False} & \mbox{False} & \mbox{False} \\
\hline
\end{array}$$

Os operadores lógicos também possuem ordem de precedência, como descrito abaixo:

$$\begin{array}{|c|c|}
\hline
\mbox{Operador} & \mbox{Associatividade} \\
\hline

\hline
** & \mbox{da direita para a esquerda}\\
\hline

\hline
*, /, //, \% &  \mbox{da esquerda para a direita} \\
\hline
+,- & \mbox{da esquerda para a direita}\\
\hline
>, <, >=, <= & \mbox{da esquerda para a direita}\\
\hline
==, != & \mbox{da esquerda para a direita} \\
\hline
\mbox{not} & \mbox{da direita para a esquerda}\\
\hline
\mbox{and} & \mbox{da esquerda para a direita} \\
\hline
\mbox{or} & \mbox{da esquerda para a direita}\\
\hline
\end{array}$$

> **Primeiro programa em** *Python*:

Agora vamos escrever uma sequência de comandos em um arquivo e escrever o nosso primeiro programa.

Instruções fundamentais em qualquer linguagem de programação são as que fazem a leitura de dados de entrada e que mostram os dados de saída.

Em *Python* temos duas funções para nos auxiliar nesta tarefa.

*   A **impressão de dados na tela** é feita usando a função print(), como argumento da função podemos ter uma mensagem entre " " ou uma variável. Por exemplo,
```
print("Olá mundo!")
```
Outro exemplo,
```
x = 2
y = 3
z = x + y
print(z)
```

*   A **leitura de dados pelo teclado** ée feita usando a função input(), como argumento da função podemos incluir uma mensagem entre " ". Por exemplo,

```
nome = input("Digite o seu nome:")
print("Seu nome é ", nome)
```
Experimente escrever um programa com as seguintes instruções:
```
x = input("Digite um número x: ")
y = input("Digite um número y: ")
z = x + y
print("O valor de ", x, " + ", y, " é: ", z)
```
O que aconteceu?

A função input devolve uma *string* (cadeia de caracteres), uma sequência de um ou mais caracteres. Se quisermos receber um número precisamos fazer uma conversão de tipo.

O operador + concatena duas *strings*.

```
x = int(input("Digite um número x: "))
y = int(input("Digite um número y: "))
z = x + y
print("O valor de ", x, " + ", y, " é: ", z)
```
**Comentários:**

As linhas de um programa escrito em *Python* iniciadas com o símbolo # são comentários. Comentários não interferem no programa, mas auxiliam os programadores a entender e documentar o código.

```
# Este programa recebe dois números inteiros
# e devolve a soma deles

x = int(input("Digite um número x: "))
y = int(input("Digite um número y: "))
z = x + y
print("O valor de ", x, " + ", y, " é: ", z)
```
